```python
# genres, movie_id, keywords, original_title, overview,cast, crew using these attributes we will create tags 
# with which we will recommend movies of similar tags 
```

```python
movies = movies[['movie_id','original_title','genres', 'keywords', 'overview','cast', 'crew']]
movies.head()
```

# Data Cleaning 

## Removing Null Values 

```python
movies.isna().sum()
```

```python
movies[movies['overview'].isna()]
```

```python
# AS the null values are present only in 3 rows we will drop it 
movies = movies.dropna()
```

## processing string input into tags format

```python
# these two attributes are in string format but we need it to be in dict format to extract name of genres 
print(type(movies['genres'][0]))
print(type(movies['keywords'][0]))
```

```python
## Converting movies names to lower case to make 
x = lambda a:a.lower()
movies['original_title'] = movies['original_title'].apply(x)
```

```python
from ast import literal_eval
```

```python
def conversion(x):
    array = np.array([])
    for i in literal_eval(x):
        array = np.append(array,i['name'])
    return array
```

```python
movies['genres'] = movies['genres'].apply(conversion)
movies.head()
```

```python
## Same process can be repeated for keywords as we are only interested in names 
```

```python
movies['keywords'] = movies['keywords'].apply(conversion)
movies.head()
```

```python
# for cast we will take first 2 actors as they would be in lead roles of the movie 
```

```python
movies['cast'][0]
```

```python
def conversion_cast(x):
    array = np.array([])
    count = 0

    for i in literal_eval(x):

        if count < 2:
            array = np.append(array, i['name'])
            count += 1

        if count == 2:
            break
    return array
```

```python
movies['cast'] = movies['cast'].apply(conversion_cast)
movies.head()
```

```python
def conversion_crew(x):
    
    for i in literal_eval(x):
        
        if (i['job']) == 'Director':
            return [i['name']]
```

```python
movies['crew']  = movies['crew'].apply(conversion_crew)
movies.head()
```

```python
# Now converting owerview into list of words to get tags 
```

```python
splitting = lambda a:(a.split()) 
movies['overview'] = movies['overview'].apply(splitting)
```

```python
movies.head()
```

```python
# Removing spaces between words in each tag 
```

```python
def space_removal(x):
    l =[]
    for i in x:
        l.append(i.replace(" ",""))    
    return l
```

```python
def space_removal_crew(x):
    l = []
    for i in x:            
        if type(i) != 'NoneType':
            l.append(i.replace(" ",""))
    return l 
```

```python
movies['cast'] = movies['cast'].apply(space_removal)
movies['genres'] = movies['genres'].apply(space_removal)
movies['keywords'] = movies['keywords'].apply(space_removal)
```

```python
movies[movies['crew'].isna()]
# there are some na values in crew table so we are droping it 
```

```python
movies.dropna(inplace=True)
movies[movies['crew'].isna()]
```

```python
movies['crew'] = movies['crew'].apply(space_removal_crew)
```

```python
## Creating tags column
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['genres'] + movies['cast'] + movies['crew'] 
movies.head()
```

```python
# removing comas and making it lower case 
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x).lower())
```

```python
movies = movies[['movie_id','original_title','tags']]
movies.head()
```

### Text Processing of tags

```python
from nltk.corpus import stopwords
from nltk import PorterStemmer

p = PorterStemmer()
```

```python
#nltk.download('stopwords')
```

```python
## Removal of stopwords
## Steming of words so that tags like 'action' or 'actions' would become one tag
```

```python
tags = movies['tags'].copy()
tags
```

```python
i = 0
for sentence in tags:
    l = []
    for word in sentence.split():
        if word not in stopwords.words('english'):
            l.append(p.stem(word))

    movies['tags'][i] = ' '.join(l)
    i = i + 1
```

```python
movies['tags']
```

```python
movies.head()
```

## Creating vectors using bag of words

```python
from sklearn.feature_extraction import text
```

```python
cv = text.CountVectorizer(max_features=10000)
vec = cv.fit_transform(movies['tags'])
```

```python
vec = vec.toarray()
```

```python
cv.get_feature_names()
```

```python
print(vec)
print('\n')
print(vec.shape)
```

```python
### In higher dimension Euclidean distance is not good so we will use Cosine Distance
### Cosine distance will gives us Theta, Smaller the theta the closer the other movie is to the 
### given movie and movie is very similar to given movie
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_vec = cosine_similarity(vec)
similarity_vec
```

```python
print(similarity_vec.shape)
```

```python
movies
```

```python
def similarity_of_movies(movie_name):
    movie_name = movie_name.lower()
    index = movies.loc[movies['original_title']==movie_name].index[0]
    
    movie_indexes = np.arange(0, similarity_vec[index].shape[0],
                         dtype='int').reshape(-1, 1)
    
    stack = np.hstack((similarity_vec[index].reshape(-1, 1),movie_indexes))
    
    sorted_stack = stack[stack[:,0].argsort()][::-1]
    
    similar_movies_index = sorted_stack[1:6][:,1]
    print(similar_movies_index)
    
    similar_movies = list(movies.iloc[similar_movies_index,1])
    
    return similar_movies
```

```python
gg = similarity_of_movies('how to train your dragon')
gg
```
